```python
#tutorial: https://www.youtube.com/watch?v=mYUyaKmvu6Y&t=970s&ab_channel=freeCodeCamp.org


import wave
import numpy as np
import matplotlib.pyplot as plt

wav_obj = wave.open('Recording.wav', 'r')

sample_freq = wav_obj.getframerate()

n_samples = wav_obj.getnframes()

t_audio = n_samples/sample_freq

signal_wave = wav_obj.readframes(n_samples)
signal_array = np.frombuffer(signal_wave, dtype=np.int16)

# for stereo:
#l_channel = signal_array[0::2]
#r_channel = signal_array[1::2]
l_signal_array = signal_array[0::2]
r_signal_array = signal_array[1::2]

times = np.linspace(0, n_samples/sample_freq, num=n_samples)

plt.figure(figsize=(15, 5))
plt.plot(times, l_signal_array)
plt.title('Audio')
plt.ylabel('Signal Value')
plt.xlabel('Time (s)')
plt.xlim(0, t_audio)
plt.show()


plt.figure(figsize=(15, 5))
plt.plot(times, r_signal_array)
plt.title('Audio')
plt.ylabel('Signal Value')
plt.xlabel('Time (s)')
plt.xlim(0, t_audio)
plt.show()
```
